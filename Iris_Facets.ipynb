```python
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,size=7, kind="reg")
```

```python
# set the hue palette as a dict for custom mapping
palette = {'setosa': "tab:cyan", 'virginica':"tab:purple", 'versicolor':"tab:pink"}

sns.jointplot(x='sepal_length',y='sepal_width', data=iris,size=7, hue="species", kind="kde",palette=palette)
```

```python
sns.stripplot(x='species',y='sepal_length',jitter=True,edgecolor='gray', data=iris,size=10,  palette=palette)
```
