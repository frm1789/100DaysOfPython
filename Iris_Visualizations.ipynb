```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


# Load the Iris dataset
iris = pd.read_csv('https://raw.githubusercontent.com/frm1789/100DaysOfPython/main/iris.data')

# Create the correlation matrix
corr = iris.corr()

# Create a heat map of the correlation matrix
plt.figure(figsize=(10,10))


#Add lines between cells:
sns.heatmap(corr, cmap="Blues", linewidth=.5,
            vmin=-1,
            annot=True)

# Add a title
plt.title('Heatmap of Iris Dataset')



# Show the plot
plt.show()
```

```python
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
plt.figure(figsize=(10,10))
sns.heatmap(corr,
            vmin=-1,
            cmap='coolwarm',
            annot=True,
            mask = mask);

# Add a title
plt.title('Heatmap of Iris Dataset')


# Show the plot
plt.show()
```
